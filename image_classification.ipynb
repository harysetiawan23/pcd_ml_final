```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df_train = pd.read_csv("train_dasets.csv",sep=',')
df_test = pd.read_csv("test_dasets.csv",sep=',')

X_train = df_train.iloc[:,1:]
y_train = df_train.iloc[:,0]

X_test = df_test.iloc[:,1:]
y_test = df_test.iloc[:,0]

df_train.shape,df_test.shape
```

```python
knn = KNeighborsClassifier(n_neighbors=5,weights='distance',p=1)
knn.fit(X_train,y_train)
```

```python
y_pred = knn.predict(X_test)
accuracy_score(y_true=y_test,y_pred=y_pred)*100
```

```python
from sklearn.svm import SVC


svc_model = SVC()
svc_model.fit(X_train,y_train)
```

```python
y_pred = svc_model.predict(X_test)
accuracy_score(y_true=y_test,y_pred=y_pred)*100
```

```python
# fig, ax = plt.subplots(figsize=(10,20)) 
# svm_report = classification_report(y_true=y_test,y_pred=y_pred, output_dict=True)
# sns.heatmap(pd.DataFrame(svm_report).iloc[:-1, :].T, annot=True,ax=ax)
print(classification_report(y_true=y_test,y_pred=y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier

tree_model = DecisionTreeClassifier()

tree_model.fit(X_train,y_train)
y_pred = tree_model.predict(X_test)
```

```python
accuracy_score(y_true=y_test,y_pred=y_pred)*100
```

```python
fig, ax = plt.subplots(figsize=(10,20)) 
decision_tree_report = classification_report(y_true=y_test,y_pred=y_pred, output_dict=True)
sns.heatmap(pd.DataFrame(decision_tree_report).iloc[:-1, :].T, annot=True,ax=ax)
```

```python
from sklearn.naive_bayes import MultinomialNB

multiNB = MultinomialNB()
multiNB.fit(X_train,y_train)

y_pred = multiNB.predict(X_test)
```

```python
accuracy_score(y_pred,y_test)*100
```

```python
fig, ax = plt.subplots(figsize=(10,20)) 
multinomial_nb_report = classification_report(y_true=y_test,y_pred=y_pred, output_dict=True)
sns.heatmap(pd.DataFrame(multinomial_nb_report).iloc[:-1, :].T, annot=True,ax=ax)
```
